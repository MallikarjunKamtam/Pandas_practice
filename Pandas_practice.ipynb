```python
import pandas as pd
```

```python
#Creating a dictionary
```

```python
df = {
    'first' : ['arjun', 'jane', 'jhon', 'kane', 'ganesh', 'rahul', 'rohan'],
    'last' : ['kamtam', 'doe', 'doe', 'watson', 'kamtam', 'yemul', 'shah'],
    'phone' : ['9550660375', '8484854946', '8049484484', '9794364946', '9875162462', '9999999999', '9898989898']
}
```

```python
df = pd.DataFrame(df)
```

```python
#converting dict into dataframe
```

```python
df
```

```python
df['full_name'] = df["first"] + ' ' + df['last']
```

```python
df.columns = [x.upper() for x in df.columns]
```

```python
df
```

```python
age = [22, 25, 19, 30, 20, 18, 25]
```

```python
# Adding new ROW
```

```python
df['age'] = age
```

```python
df
```

```python
#sorting by last name firstly and first name secondly
```

```python
df.sort_values(by = ['LAST', 'FIRST'], ascending = [False, True])
```

```python
#Sorting by age permenantly
```

```python
df
```

```python
df.sort_values(by = 'age', ascending = True, inplace = True)
```
